## K-means Clustering to define mood for each song


### 1. Find the Optimal K

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
```

```python
music_df.head(5)
```

```python
k_means_df = music_df[['valence']]
k_means_df = k_means_df.dropna()
```

```python
k_means_df.head(5)
```

```python
# #normalize key, loudness, tempo, duration_ms, time_signature

# loudness = k_means_df[['loudness']].values
# min_max_scaler = preprocessing.MinMaxScaler()
# loudness_scaled = min_max_scaler.fit_transform(loudness)

# tempo = k_means_df[['tempo']].values
# min_max_scaler = preprocessing.MinMaxScaler()
# tempo_scaled = min_max_scaler.fit_transform(tempo)

# duration_ms = k_means_df[['duration_ms']].values
# min_max_scaler = preprocessing.MinMaxScaler()
# duration_scaled = min_max_scaler.fit_transform(duration_ms)

# time_signature = k_means_df[['time_signature']].values
# min_max_scaler = preprocessing.MinMaxScaler()
# time_scaled = min_max_scaler.fit_transform(time_signature)
```

```python
from sklearn.cluster import KMeans
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(k_means_df)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
distance_df = pd.DataFrame(Sum_of_squared_distances)
distance_df = distance_df.rename(columns={0:"Sum_of_squared_distances"})
distance_df
```

```python
K_df = pd.DataFrame(K).astype(str)
K_df = K_df.rename(columns={0:"K_Value"})
K_df 
```

```python
knn_df = pd.concat([K_df, distance_df], axis=1)
knn_df
```

```python
# As we can see, when k = 5 the average squared distance to cluster centre starts to trail off significantly. 
# This is the long awaited elbow, our optimal k.
knn_df.plot(x ='K_Value', y='Sum_of_squared_distances', kind = 'line',title="Elbow Method for Optimal K",style='.-')
```
